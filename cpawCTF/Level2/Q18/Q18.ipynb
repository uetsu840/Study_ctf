```python
import re

def nchars(s, n):
    """文字列 s に、同じ文字が n 個以上連続している部分文字列を見つける
    """
    assert n > 0
    reg = re.compile("(.)\\1{%d,}" % (n - 1))  # カンマを取ると n 個ちょうどになる
    while True:
        m = reg.search(s)
        if not m:
            break
        yield m.group(0)
        s = s[m.end():]


flag_seed = list(nchars(lines[0], 3))
```

```python
ans = ""
for chars in flag_seed:
    ans = ans + chars[0]
print(ans.lower())
```
